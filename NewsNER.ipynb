```python
from transformers import pipeline

nlp = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", tokenizer="bert-base-cased")

def extract_entities(sentences):
    entities = []
    for sentence in sentences:
        result = nlp(sentence)
        entities.append(result)
    return entities

sentences = ["The President of the United States met with the Prime Minister of the United Kingdom in London.", 
             "China and India have agreed to a ceasefire along the disputed border.", 
             "North Korea has conducted another test of its nuclear missile program."]

entities = extract_entities(sentences)
print(entities)
```
